```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
datafile = pd.read_csv(r"D:\PDFs\GIS & Remote Sensing\NASA International Space Challenge 2024\Banglore_traffic_Dataset.csv")
datafile.head(10)
```

```python
datafile.columns
```

```python

# Split the Coordinates column into Latitude and Longitude
datafile[['Latitude', 'Longitude']] = datafile['Coordinates'].str.split(',', expand=True)
datafile['Latitude'] = pd.to_numeric(datafile['Latitude'], errors='coerce')
datafile['Longitude'] = pd.to_numeric(datafile['Longitude'], errors='coerce')
```

```python
datafile.head(10)
```

```python
area_locations = {
    'Indiranagar': [12.9762459,77.6057284],
    'Whitefield': [12.9645675,77.7365327],
    'Koramangala': [12.9354978,77.6115928],
    'M.G.Road': [12.9747358,77.6069128],
    'Jayanaagr': [12.9292533,77.5665553],
    'Hebbel': [13.034661,77.559204],
    'Yeshwanthpur': [13.027129,77.5251987],
    'Electronic City': [12.8380004,77.6329978]
}

# Create a base map centered on Bengaluru
bengaluru_map = folium.Map(location=[12.9544595,77.3012695], zoom_start=12)

# Prepare heat data
heat_data = []

for index, row in datafile.iterrows():
    area_name = row['Area Name']
    
    if area_name in area_locations:
        lat, lon = area_locations[area_name]
        congestion_level = row['Congestion Level'] / 100  # Scale as needed
        heat_data.append([lat, lon, congestion_level])

# Add HeatMap if data is available
if heat_data:
    HeatMap(heat_data).add_to(bengaluru_map)
    display(bengaluru_map)
else:
    print("No heatmap data available! Please check your dataset and area mappings.")
```
